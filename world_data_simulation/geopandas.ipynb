```python
import geopandas as gpd
world = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))
world
```

```python
world = gpd.read_file("data/ne_10m_admin_0_countries/ne_10m_admin_0_countries.shp")
world
```

```python
names_list = world['NAME_EN'].tolist()
names_list
```

```python
# Plot by continent
world.plot(column="CONTINENT", legend=True, figsize=(12, 6))
```

```python
ax = world.plot(figsize=(12,6), color="lightgrey", edgecolor="black")
ax.set_title("World Base Map", fontsize=14)
ax.set_axis_off()
plt.show()
```

```python
# ==========================================
# 3. Example: Merging data by country
# ==========================================
# Fake dataset
df = pd.DataFrame({
    "name": ["France", "Germany", "United States of America", "Brazil"],
    "value": [5, 15, 8, 20]
})

# Merge with world polygons
world_data = world.merge(df, left_on="NAME_EN", right_on="name", how="left")

ax = world_data.plot(
    column="value",
    legend=True,
    figsize=(12,6),
    missing_kwds={"color": "lightgrey", "label": "No data"}
)
ax.set_title("Choropleth Example", fontsize=14)
ax.set_axis_off()
plt.show()
```

```python
# ==========================================
# 4. Example: Plotting points with coordinates
# ==========================================
points_df = pd.DataFrame({
    "city": ["Paris", "London", "Berlin", "New York", "Tokyo", "Mecca", "Sydney", "Rio de Janeiro", "Brasilia", "Washington D.C.", "Toronto", "Beijing", "Singapore"],
    "lon": [2.3514, -0.1278, 13.4050, -74.0060, 139.650, 39.8173, 151.2057, -43.1729, -47.8919, -77.0369, -79.383, 116.4074, 103.8198],
    "lat": [48.8575, 51.5074, 52.5200, 40.7128, 35.6764, 21.4241, -33.8727, -22.9068, -15.7975, 38.9072, 43.6532, 39.9042, 1.3521],
    "measure": [5.1, 3.2, 7.7, 4.5, 10, 100, 2.7, 1.6, 2.3, 4.9, 5.3, 7.6, 10]
})

points_gdf = gpd.GeoDataFrame(
    points_df,
    geometry=gpd.points_from_xy(points_df["lon"], points_df["lat"]),
    crs="EPSG:4326"
)

ax = world.plot(figsize=(12,6), color="white", edgecolor="grey")
points_gdf.plot(ax=ax, markersize=10, color="red")
ax.set_title("Points Example", fontsize=14)
ax.set_axis_off()
plt.show()
```
